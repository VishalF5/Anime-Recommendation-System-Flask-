```python
df=pd.read_csv("/content/drive/MyDrive/Anime Planet.csv")
```

```python
df.head(2)
```

```python
index=[]
for i in range (0,len(df)):
    per=df.iloc[i,].isna().sum()/df.shape[1]*100
    index.append(per)
    
```

```python
ind=[]
for i,u in enumerate(index):
    if u>70:
        ind.append(i)
        
```

```python
df=df.drop(df.index[ind])
```

```python
df.columns
```

```python
df = df.loc[:,['Gener','Demographic','Studio','Synopsis','Theme']]
```

```python
df.head()
```

```python
df['Gener'] = df['Gener'].replace(np.nan, 'unknown')
df['Demographic'] = df['Demographic'].replace(np.nan, 'unknown')
df['Synopsis'] = df['Synopsis'].replace(np.nan, 'unknown')
df['Studio'] = df['Studio'].replace(np.nan, 'unknown')
df['Theme'] = df['Theme'].replace(np.nan, 'unknown')
```

```python
genre=[]
for i in df['Gener']:
    genre.append(re.sub(',','', i))
```

```python
df['Gener']=genre
```

```python
syn=[]
for i in df['Synopsis']:
    syn.append(re.sub(r'[,\.\n\r\'\"\[\]\,]','', i))
```

```python
df['Synopsis']=syn
```

```python
Theme=[]
for i in df['Theme']:
    Theme.append(re.sub(r'[,\.\n\r\'\"\[\]\,]','', i))
    
df['Theme']=Theme
```

```python
df
```

```python
df['Comb']=df['Gener']+' '+df['Demographic']+' '+df['Studio']+' '+df['Synopsis']+' '+df['Theme']
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(df['Comb']).toarray()
```

```python
vector.shape
```

```python
similarity = cosine_similarity(vector)
```

```python
import pickle
```

```python
pickle.dump(similarity.to_dict, open('similarity.pickle', 'wb'))
```

```python
def recommend(anime):
    index = df[df['Name'] == anime].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:5]:
        print(df.iloc[i[0]].Name)
```
